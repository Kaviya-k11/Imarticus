## PCA Analysis Using Iris Dataset

```python
import sklearn
from sklearn.datasets import load_iris
```

```python
df=load_iris()
df
```

```python
x=df.data
x
 
```

```python
y=df.target
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.60,test_size=0.40,random_state=0)

```

### classifictaiom

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

### Accuracy

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

### Improve Accuracy Using PCA

```python
 
from sklearn.decomposition import PCA

pca=PCA(n_components=0.95)
pca
```

```python
# pca only done for x not for y

pca.fit(x_train)
```

```python
x_train_transformed=pca.transform(x_train)
print(x_train_transformed.shape)
```

```python
x_train_transformed
```

```python
x_test_transformed=pca.transform(x_test)
print(x_test_transformed.shape)
```

```python
x_test_transformed
```

```python
new_model=LogisticRegression(penalty='l1',solver='liblinear')

new_model
```

```python
new_model.fit(x_train_transformed,y_train)
```

```python
new_prediction=new_model.predict(x_test_transformed)
new_prediction
```

```python
from sklearn.metrics import accuracy_score

new_accuracy=accuracy_score(new_prediction,y_test)
new_accuracy
```
