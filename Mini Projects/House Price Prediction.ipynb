```python
for i in ['MSSubClass','OverallQual','OverallCond']:
    df[i] = df[i].astype('object')
    print(df[i].dtypes)
```

```python
import datetime as dt
current_year = int(dt.datetime.now().year) 
```

```python
Buiding_age = current_year - df.YearBuilt
Remodel_age = current_year - df.YearRemodAdd
```

```python
df['Buiding_age'] = Buiding_age
df['Remodel_age'] = Remodel_age
```

```python
df.shape
```

```python
df['LotFrontage'].fillna(df['LotFrontage'].median() , inplace = True)
df['MasVnrArea'].fillna(0 , inplace = True)
df['GarageYrBlt'].fillna(0 , inplace = True)
```

```python
for col in ['BsmtQual','BsmtCond','BsmtExposure','BsmtFinType1','BsmtFinType2']:
    df[col].fillna('No Basement' , inplace = True)
df['MasVnrType'].fillna('None' , inplace = True)
df['Electrical'].fillna('SBrkr' , inplace = True)
df['FireplaceQu'].fillna('No Fireplace' , inplace = True)
for col in ['GarageType','GarageFinish','GarageQual','GarageCond']:
    df[col].fillna('No Garage' , inplace = True)
```

```python
df['PoolQC'].fillna('No Pool' , inplace = True)
df['Fence'].fillna('No Fence' , inplace = True)
df['MiscFeature'].fillna('None' , inplace = True)
df['Alley'].fillna('No alley access' , inplace = True)
```

```python
df1 =df.select_dtypes(include=[np.number])
```

```python
x=df1.drop('SalePrice',axis=1)
y=df1['SalePrice']
```

```python
from sklearn.preprocessing import StandardScaler
df1= StandardScaler().fit_transform(df1)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression().fit(xtrain,ytrain)
```

```python
ypred=model.predict(xtest)
```

```python
from sklearn.metrics import r2_score
print(r2_score(ypred,ytest))
```
